## **Question 2:** Logistic Regression

Recall that the hypothesis/decision rule in a logistic regression model is given by</p>

$$h_w(x) = \sigma(w^Tx) \\ \text{where } \sigma  \text{ is the sigmoid function}$$

<p>Since logistic regression does not have a closed form solution, we will use gradient descent to obtain the parameters $w$. We will use the negative log likelihood loss with L2 regularization as the loss function. Mathematically, the loss function $l(w)$ for a given set of parameters $w$ will be,</p>

$$l(w) = NLL(w) + \frac{\lambda}{2}||w||^2 \\ \text{where } NLL(w) = -\frac{1}{n} \sum_{i=1}^{n} y_i\log(h(x_i)) + (1 - y_i)\log(1 - h(x_i))$$



### **2.1**

Compute the derivate $\nabla_w l(w)$. 

**Hints:** 
- if $h(x) = f(x) + g(x)$, then $\nabla_x h(x)$ = $\nabla_x f(x)$ + $\nabla_x g(x)$

- $[\frac{f(x)}{g(x)}]'$ = $\frac{g(x)f'(x) - f(x)g'(x)}{(g(x))^2}$

- $[f(g(x)]'$ = $f'(g(x))$ * $g(x)'$ 



**Answer**

$$l(w) = NLL(w) + \frac{\lambda}{2}||w||^2 $$ \\

Start by substituting $NLL(w)$ \\


$$l(w) = -\frac{1}{n} \sum_{i=1}^{n} y_i\log(h(x_i)) + (1 - y_i)\log(1 - h(x_i)) + \frac{\lambda}{2}||w||^2 $$

Substitute in $h_w(x) = \sigma(w^Tx)$

$$l(w) = -\frac{1}{n} \sum_{i=1}^{n} y_i\log(\sigma(w^Tx_i))) + (1 - y_i)\log(1 - \sigma(w^Tx_i)) + \frac{\lambda}{2}||w||^2 $$

Start the derivative

$$\nabla_w l(w) = -\frac{1}{n} \sum_{i=1}^{n} \frac{y_i}{\sigma(w^Tx_i)} * \nabla_w \sigma(w^Tx_i)  + \frac{1 - y_i}{1 - \sigma(w^Tx_i)}* \nabla_w (1 - \sigma(w^Tx_i)) + \lambda * w $$

Next, keep the following in mind:

$$ \sigma(x) = \frac{1}{1 + e^{-x}} $$ \\

So, using quotient rule

$$\nabla_x \sigma(x) = \frac{0 * (1 + e^{-x}) - 1 * (e^{-x} * (-1))}{(1 + e^{-x})^2}$$

$$\nabla_x \sigma(x) = \frac{e^{-x}}{(1 + e^{-x})^2} = \frac{1 - 1 + e^{-x}}{(1 + e^{-x})^2}  = \frac{1 + e^{-x}}{(1 + e^{-x})^2} + \frac{-1}{(1 + e^{-x})^2} $$

$$\nabla_x \sigma(x) = \frac{1}{1 + e^{-x}} * (1 - \frac{1}{1 + e^{-x}}) $$

$$\nabla_x \sigma(x) = \sigma(x) * (1 - \sigma(x)) $$

Now, substitute the above into our derivative

$$\nabla_w l(w) = -\frac{1}{n} \sum_{i=1}^{n} \frac{y_i}{\sigma(w^Tx_i)} * (\sigma(w^Tx_i) * (1 - \sigma(w^Tx_i))) * x_{ij} + \frac{1 - y_i}{1 - \sigma(w^Tx_i)}* (- \sigma(w^Tx_i) * (1 - \sigma(w^Tx_i))) * x_{ij} + \lambda * w $$ 

$$\nabla_w l(w) = -\frac{1}{n}\sum_{i=1}^{n} y_i * (1 - \sigma(w^Tx_i)) * x_{ij} - (1 - y_i) * \sigma(w^Tx_i) * x_{ij} + \lambda * w $$ 

$$\nabla_w l(w) = -\frac{1}{n}\sum_{i=1}^{n} y_i * (1 - h_w(x_i)) * x_{ij} - (1 - y_i) * h_w(x_i) * x_{ij} + \lambda * w $$ 

Take $x_ij$ as a common factor

$$\nabla_w l(w) = -\frac{1}{n}\sum_{i=1}^{n} [y_i * (1 - h_w(x_i)) - (1 - y_i) * h_w(x_i)] * x_{ij} + \lambda * w $$ 

Expand all the internal parts

$$\nabla_w l(w) = -\frac{1}{n}\sum_{i=1}^{n} [y_i - y_i *h_w(x_i)) - h_w(x_i) + y_i * h_w(x_i)] * x_{ij} + \lambda * w $$ 

Cancel out $y_i *h_w(x_i)$

$$\nabla_w l(w) = -\frac{1}{n}\sum_{i=1}^{n} [y_i - h_w(x_i)] * x_{ij} + \lambda * w $$

Finally, rewite it into matrix form


$$\nabla_w l(w) = (\hat{y} - y)^{T} \cdot X + \lambda \cdot w$$
